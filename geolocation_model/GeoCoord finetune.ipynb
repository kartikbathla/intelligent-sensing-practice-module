```python
def get_datetime_axis(
        image_url: str
) -> dict:
    response = client.responses.create(
        model="gpt-4.1-mini",
        input=[
            {
                "role": "user",
                "content": [
                    {
                        "type": "input_text",
                        "text": '''
                        Instructions: Find the time and the name stated after "axis-" in the image.
                        Give me the time in HH:MM:SS MM/DD/YYYY format only

                        Return me in this format:
                        {datetime:"", axis:""}

                        ''',
                    },
                    {
                        "type": "input_image",
                        "image_url": f"{image_url}",
                    }
                ],
            }
        ]
    )

    match = re.search(r"{datetime:\"(.*?)\",\s*axis:\"(.*?)\"}", response.output_text)

    if match:
        datetime_str = match.group(1)
        axis_str = match.group(2)
        namespacedict = {"datetime": datetime_str, "axis": axis_str}
    else:
        print("Error: Datetime or axis not found in response.")
        namespacedict = {}  # Return an empty dictionary if no match

    return namespacedict
```

```python
def get_image_id(image_path):
   """Retrieves the Google Drive ID of an image file.

   Args:
     image_path: The path to the image file in your Google Drive.

   Returns:
     The Google Drive ID of the image file.
   """
   try:
     results = drive_service.files().list(q=f"name='{os.path.basename(image_path)}' and trashed=false", fields="nextPageToken, files(id, name)").execute()
     items = results.get('files', [])
     if items:
       return items[0]['id']
     else:
       return None  # Image not found
   except Exception as e:
     print(f"An error occurred: {e}")
     return None
```

```python
def get_camera_json(request_timestamp):
    URL = 'https://cameras.alertcalifornia.org/public-camera-data/all_cameras-v3.json?rqts='

    # Format timestamp to UNIX
    unix_timestamp = int(request_timestamp.timestamp())
    URL += str(unix_timestamp)

    reponse = requests.get(URL)
    return reponse.json()
```

# Model Architect

         +------------------------+
         |   Fire image patch     |  -> CNN backbone -> feature vector (512d)
         +------------------------+

         +------------------------+
         |  Camera metadata (FOV, |  -> Fully connected layers -> vector (128d)
         |  azimuth, tilt, etc.)  |
         +------------------------+

                      
            Concatenate features
                      
              Fully connected head
                      
              Predict [lat, lon]

```python
import tensorflow as tf
from tensorflow.keras import layers, Model
from tensorflow.keras.applications import MobileNetV2

image_input = tf.keras.Input(shape=(224, 224, 3), name='image_input')

# Import mobilenetv2 here for embeddings
base_model = MobileNetV2(
    input_shape=(224, 224, 3),
    include_top=False,
    weights='imagenet'
)
base_model.trainable = False
x = base_model(image_input)
x = layers.GlobalAveragePooling2D()(x)

# Metadata input:
# [camera_lat, camera_lon, camera_alt, # Camera location information
#  fov_left_azimuth, fov_right_azimuth, # Camera FOV
#  tilt_angle, zoom_level, # Camera lens information
#  bbox_x_center_norm, bbox_y_center_norm, bbox_area_norm] # Bounding boxes from YoloV8

meta_input = tf.keras.Input(shape=(19,), name='meta_input')  # shape=(batch_size, 10)
meta_output = layers.Dense(64, activation='relu')(meta_input)
meta_output = layers.Dense(32, activation='relu')(meta_output)

# Combine both branches
combined = layers.Concatenate()([x, meta_output])
z = layers.Dense(128, activation='relu')(combined)
z = layers.Dense(64, activation='relu')(z)

# Output: Predict lat/lon
output = layers.Dense(2, activation='linear', name='geo_output')(z)

# Final model
fusion_model = Model(inputs=[image_input, meta_input], outputs=output)
fusion_model.summary()
```

## Comparing to manual calculations

```python
# Credit: Tong Chang
def compute_bbox_geo_coordinates(
    img_width, img_height, fov_h, camera_altitude, ground_altitude,
    fov_lft, fov_rt, bbox_x, bbox_y, bbox_width, bbox_height, ground_ratio=0.8
):
    """
    
    """
    # 
    h = camera_altitude - ground_altitude

    # 
    ground_width = 2 * h * np.tan(np.radians(fov_h / 2))
    ground_height = (img_height / img_width) * ground_width

    # 
    lon_range = fov_rt[0] - fov_lft[0]
    lat_range = fov_lft[1] - fov_rt[1]

    # 
    bbox_center_x = bbox_x  # 
    bbox_center_y = bbox_y + bbox_height / 2

    # 
    ground_y_start = img_height * (1 - ground_ratio)
    ground_y_height = img_height * ground_ratio

    # 
    lon_ratio = bbox_center_x / img_width
    lat_ratio = (bbox_center_y - ground_y_start) / ground_y_height

    # 
    bbox_left_lon = fov_lft[0] + lon_ratio * lon_range
    bbox_left_lat = fov_lft[1] - lat_ratio * lat_range

    return bbox_left_lon, bbox_left_lat
```

```python
from PIL import Image
import os

# Assuming image_files is aligned with X and sorted accordingly
image_files = os.listdir(image_folder_path)

# Filter image_files using the none_indices, to align with filtered X
filtered_image_files = [
    img for i, img in enumerate(image_files) if i not in none_indices
]

geo_coords = []
ground_altitude = 0  # Assume 0 if unknown

for i, (x, image_filename) in enumerate(zip(X, filtered_image_files)):
    try:
        image_path = os.path.join(image_folder_path, image_filename)
        with Image.open(image_path) as img:
            img_width, img_height = img.size

        # unpack your data from X
        fov_lft = [x[0], x[1]]  # [lon, lat]
        fov_rt = [x[2], x[3]]  # [lon, lat]
        fov_h = x[6]  # horizontal FOV
        azimuth = x[7]
        tilt = x[8]
        zoom = x[9]
        camera_lon = x[10]
        camera_lat = x[11]
        camera_altitude = x[12]

        bbox_x = x[13]
        bbox_y = x[14]
        bbox_width = x[15]
        bbox_height = x[16]

        lon, lat = compute_bbox_geo_coordinates(
            img_width, img_height, fov_h, camera_altitude, ground_altitude,
            fov_lft, fov_rt, bbox_x, bbox_y, bbox_width, bbox_height
        )

        geo_coords.append((lon, lat))
    except Exception as e:
        print(f"Error at index {i} ({image_filename}): {e}")
        geo_coords.append(None)
```

```python
import json
import os

def rename_json_keys_strip_zeros(json_path, save_to=None):
    """
    Loads a JSON file and renames keys by removing leading zeros from the numbered suffixes in image filenames.

    Args:
        json_path (str): Path to the original JSON file.
        save_to (str, optional): Path to save the updated JSON file. If None, the file is not saved.

    Returns:
        dict: Updated dictionary with renamed keys.
    """
    with open(json_path, 'r') as f:
        data = json.load(f)

    updated_data = {}

    for key, value in data.items():
        name, ext = os.path.splitext(key)
        if "_" in name:
            prefix, number = name.rsplit("_", 1)
            if number.isdigit():
                new_number = str(int(number))  # remove leading zeros
                new_key = f"{prefix}_{new_number}{ext}"
            else:
                new_key = key  # unchanged
        else:
            new_key = key  # unchanged

        updated_data[new_key] = value

    if save_to:
        with open(save_to, 'w') as f:
            json.dump(updated_data, f, indent=2)
        print(f"Updated JSON saved to: {save_to}")

    return updated_data
```

```python
import os
import json
from PIL import Image

def construct_test_X_y(
    image_dir: str,
    image_files: list,
    image_names: list,
    cameras_information: list,
    ground_truth_df,
    annotations: str
):
    """
    Constructs X and y from fully pre-filtered image files, names, and metadata.

    Args:
        image_dir (str): Path to directory with test images.
        image_files (list): List of filenames (e.g. ['foo_fire_3.jpg']).
        image_names (list): Cleaned fire names (e.g. ['Foo Fire']).
        cameras_information (list): List of metadata dicts (from get_camera_json).
        ground_truth_df (pd.DataFrame): DF with 'incident_name', 'incident_longitude', 'incident_latitude'.
        annotations_path (str): Path to annotations JSON.

    Returns:
        Tuple: (X, y)
    """

    X, y = [], []

    camera_keys = [
        "fov_lft", "fov_rt", "fov_center",
        "fov", "az_current", "tilt_current", "zoom_current"
    ]

    for file, fire_name, cam in zip(image_files, image_names, cameras_information):
        props = cam["properties"]
        coords = cam["geometry"]["coordinates"]

        # 1. Camera metadata
        x_vector = []
        for key in camera_keys:
            val = props.get(key)
            if isinstance(val, list):
                x_vector.extend(val)
            else:
                x_vector.append(val)
        x_vector.extend(coords)  # cam lon, lat, alt

        # 2. Bounding box
        if file not in annotations:
            print(f"Skipping {file}  missing annotation")
            continue
        bbox = annotations[file][0]
        x_vector += [bbox["x"], bbox["y"], bbox["width"], bbox["height"]]

        # 3. Image dimensions
        img_path = os.path.join(image_dir, file)
        try:
            with Image.open(img_path) as img:
                w, h = img.size
                x_vector += [w, h]
        except:
            print(f"Skipping {file}  image not found")
            continue

        # 4. Ground truth coordinates
        match = ground_truth_df[ground_truth_df["incident_name"] == fire_name]
        if match.empty:
            print(f"Skipping {file}  no ground truth match for {fire_name}")
            continue
        y_vector = match[["incident_longitude", "incident_latitude"]].values[-1]

        # Add to final dataset
        X.append(x_vector)
        y.append(y_vector)

    return X, y
```

```python
import numpy as np

def evaluate_model_and_geocalc_mae(model, X_dict, y_true, compute_bbox_geo_coordinates, ground_altitude=0):
    """
    Evaluates model prediction MAE and calculated geo-coordinate MAE.

    Args:
        model: Trained Keras model.
        X_dict: Dictionary with 'image_input' and 'meta_input'.
        y_true: Ground truth coordinates (list of [lon, lat]).
        compute_bbox_geo_coordinates: Function to compute (lon, lat) from metadata.
        ground_altitude: Ground reference altitude (default is 0).

    Returns:
        Dictionary with MAEs for model predictions and geo coordinate calculation.
    """

    # Step 1: Model Prediction MAE
    predictions = model.predict(X_dict)
    y_array = np.array(y_true)
    pred_array = np.array(predictions)

    mae_pred_lat = np.mean(np.abs(pred_array[:, 1] - y_array[:, 1]))
    mae_pred_lon = np.mean(np.abs(pred_array[:, 0] - y_array[:, 0]))

    # Step 2: Geo Coordinate Calculation MAE
    geo_coords = []
    y_cleaned = []

    for x_meta, target in zip(X_dict["meta_input"], y_true):
        try:
            fov_lft = [x_meta[0], x_meta[1]]
            fov_rt = [x_meta[2], x_meta[3]]
            fov_h = x_meta[6]
            camera_altitude = x_meta[12]
            bbox_x = x_meta[13]
            bbox_y = x_meta[14]
            bbox_width = x_meta[15]
            bbox_height = x_meta[16]
            img_width = x_meta[-2]
            img_height = x_meta[-1]

            lon, lat = compute_bbox_geo_coordinates(
                img_width, img_height, fov_h, camera_altitude, ground_altitude,
                fov_lft, fov_rt, bbox_x, bbox_y, bbox_width, bbox_height
            )

            geo_coords.append([lon, lat])
            y_cleaned.append(target)

        except Exception as e:
            print(f"Error processing data: {e}")
            continue

    geo_array = np.array(geo_coords)
    y_array_cleaned = np.array(y_cleaned)

    mae_geo_lat = np.mean(np.abs(geo_array[:, 1] - y_array_cleaned[:, 1]))
    mae_geo_lon = np.mean(np.abs(geo_array[:, 0] - y_array_cleaned[:, 0]))

    # Results
    return {
        "model_mae": {"latitude": mae_pred_lat, "longitude": mae_pred_lon},
        "geo_calc_mae": {"latitude": mae_geo_lat, "longitude": mae_geo_lon}
    }
```

```python
import tensorflow as tf
from tensorflow.keras import layers, Model
from tensorflow.keras.applications import MobileNetV2

image_input = tf.keras.Input(shape=(224, 224, 3), name='image_input')

# Import mobilenetv2 here for embeddings
base_model = MobileNetV2(
    input_shape=(224, 224, 3),
    include_top=False,
    weights='imagenet'
)
base_model.trainable = False
x = base_model(image_input)
x = layers.GlobalAveragePooling2D()(x)

# Metadata input:
# [camera_lat, camera_lon, camera_alt, # Camera location information
#  fov_left_azimuth, fov_right_azimuth, # Camera FOV
#  tilt_angle, zoom_level, # Camera lens information
#  bbox_x_center_norm, bbox_y_center_norm, bbox_area_norm] # Bounding boxes from YoloV8

meta_input = tf.keras.Input(shape=(19,), name='meta_input')  # shape=(batch_size, 10)
meta_output = layers.Dense(64, activation='relu')(meta_input)
meta_output = layers.Dense(32, activation='relu')(meta_output)

# Combine both branches
combined = layers.Concatenate()([x, meta_output])
z = layers.Dense(128, activation='relu')(combined)
z = layers.Dense(64, activation='relu')(z)

# Output: Predict lat/lon
output = layers.Dense(2, activation='linear', name='geo_output')(z)

# Final model
fusion_model = Model(inputs=[image_input, meta_input], outputs=output)
fusion_model.summary()
```
